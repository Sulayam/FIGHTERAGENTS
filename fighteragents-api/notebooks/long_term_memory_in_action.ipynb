```python
import sys
sys.path.insert(0, "../src")

from langchain_core.documents import Document
from fighteragents.application import LongTermMemoryRetriever
```

```python
from fighteragents.config import settings

# Override MongoDB connection string
settings.MONGO_URI = (
    "mongodb://fighteragents:fighteragents@localhost:27017/?directConnection=true"
)
```

```python
def print_memories(memories: list[Document]) -> None:
    for i, memory in enumerate(memories):
        print("-" * 100)
        print(f"Memory {i + 1}:")
        print(f"{i + 1}. {memory.page_content[:100]}")
        print(f"Source: {memory.metadata['source']}")
        print("-" * 100)
```

```python
retriever = LongTermMemoryRetriever.build_from_settings()

memories = retriever("Joe Rogan")
print_memories(memories)
```

```python
memories = retriever("Khamzat Chimaev")
print_memories(memories)
```
